```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
imagen = cv2.imread('PSI_Tray031_2015-12-27--09-03-57_top.png')
```

```python
# Convert the image from BGR to RGB (Matplotlib uses RGB)
imagen_rgb = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)

# Modify the image by manipulating RGB values
modified_image = imagen_rgb.copy()  # Create a copy to modify

# Check and modify green values
for i in range(modified_image.shape[0]):
    for j in range(modified_image.shape[1]):
        # Check if the green value is greater than 80
        if modified_image[i, j, 1] <= 90:  # Green component
            # If not, set the pixel to black (0, 0, 0)
            modified_image[i, j] = [0, 0, 0]

# Display the original and modified image
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(imagen_rgb)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Modified Image')
plt.imshow(modified_image)
plt.axis('off')

plt.tight_layout()
plt.show()
```
